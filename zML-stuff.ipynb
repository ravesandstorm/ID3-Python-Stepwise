```python
import math
def log(x): return math.log(x)/math.log(2)
def entropy(x, y=None):
    if y is None:
        ent = 0
        tot = x.count()
        for i in x.unique():
            num = list(x).count(i)
            ent -= num/tot*log(num/tot)
            #print('es / =',num, tot)
        return ent
    else:
        ents = []
        for i in x.unique():
            inds=np.where(x == i)[0]
            den = list(x).count(i)
            vals = y[inds]
            nums = [list(vals).count(i) for i in vals.unique()]
            print(x.name,':',i, nums, den)
            ents += [sum([-i*log(i) for i in [i/den for i in nums]])]
        return ents

x = df.iloc[:, 0]
y = df.iloc[:, -1]

es = df.iloc[:, -1]
es = entropy(es)
ents = entropy(x,y)
print('es =',es)
ents, x.name
```

```python
for v in range(df.shape[1]-1):
    x = df.iloc[:, v]
    y = df.iloc[:, -1]
    ents = entropy(x,y)
    fracs = [list(x).count(i) for i in x.unique()]
    IG = es
    for j,i in enumerate(fracs):
        IG -= i/y.count() * ents[j]
        print(i, y.count(), ents[j])
        print(f'Entropy for {x.unique()[j]} = {ents[j]:.4f}')
    print(f'-----------------IG for {x.name} = {IG:.4f}')
```
